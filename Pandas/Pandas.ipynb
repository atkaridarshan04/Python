# drop_duplicates

```python
df7 = pd.DataFrame({"name": ['Batman', 'Batman', 'Catwoman'],
                   "toy": ['Superman', 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
df7
```

```python
df7.drop_duplicates(subset=['name']) 
```

```python
df7.drop_duplicates(subset=['name'], keep='last') #by default keep='first'
```

```python
df7.drop_duplicates(subset=['name'], keep=False) # drops all which are duplicate
```

# Data Visualisation

```python
# pandas have built-in matplot library
import pandas as pd
import numpy as np
%matplotlib inline

dv = pd.DataFrame(np.random.randn(5,5), index =[10,20,30,40,50], columns=['A','B','C','D','E'] )

dv = dv.cumsum()
dv.plot()
```

```python
ts = pd.Series(np.random.randn(5), index=[1,2,3,4,5])
ts = ts.cumsum()
ts.plot()
```
