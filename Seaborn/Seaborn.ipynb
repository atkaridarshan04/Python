# Barplot

```python
# res = sns.barplot(x = 'cyl', y ='mpg', data = mtcars) # basic to make a barplot graph

res = sns.barplot(x = 'cyl', y ='mpg', data = mtcars, hue='cyl', legend=False, palette = 'Set2')
# palette set the color
# hue refers to the color encoding of data points in a plot. It is one of the main properties of a color

plt.title("mpg vs cyl")
plt.show()
```

# Countplot
A Countplot in Seaborn is a graphical display that shows the counts of observations in each categorical bin using bars

```python
sns.countplot(x='carb',data = mtcars, hue = 'cyl', legend = False, palette = 'crest')
# cannot pass both x and y at same time, as countplot gives the count of data
```

### Horizontal Countplot

```python
sns.countplot(y='gear', data = mtcars, hue='cyl')
```

# Histplot
It is a classic visualization tool that represents the distribution of the data by counting the number of observations that fall within discrete bins. The plot displays the count of observations in each bin as a bar.

```python
sns.histplot(mtcars.mpg, bins= 10)
```

# HeatMaps
A heatmap in Seaborn is a graphical representation of data using colors to visualize the value of a matrix.

```python
numeric_mtcars = mtcars.select_dtypes(include=['float64', 'int64'])

sns.heatmap(numeric_mtcars.corr(), cbar = True, linewidths = 0.5)
# cbar = color bar
```

```python
iris = sns.load_dataset('iris') # loading inbuild dataset in seaborn
iris.head()
```

```python
iris['species'].value_counts()
```

# Scatterplot
A scatter plot in seaborn is a graph that shows the relationship between two sets of continuous numerical data. Its a powerful tool for visualizing the relationship between two variables. 

```python
# sns.scatterplot(x='sepal_length', y='petal_length', data = iris)
sns.scatterplot(x='sepal_length', y='petal_length', data = iris, hue='species', size='species')
```

# Pairplot
Pairplot is a visualization technique used to plot pairwise relationships in a dataset. It is a high-level interface for PairGrid, which is a module in the seaborn library. 

```python
# sns.pairplot(iris)
sns.pairplot(iris, hue = 'species', palette='Set1')
```

# lmplot
lmplot is a figure-level function that combines the functionality of regplot and FacetGrid. It is used to draw a scatter plot onto a FacetGrid, which is a grid of subplots. The lmplot function is intended as a convenient interface to fit regression models across conditional subsets of a dataset.

```python
# sns.lmplot(x='petal_length', y='petal_width', data = iris)
sns.lmplot(x='petal_length', y='petal_width', data = iris, hue='species', markers=["*","^",'o'])
```

# Boxplot
Boxplots, also known as box plots or box-and-whisker plots, are a standardized way of displaying the distribution of a data set based on its five-number summary of data points: the minimum, first quartile (Q1), median, third quartile (Q3), and maximum. They provide a visual representation of the data distribution, helping to identify the range, skewness, and outliers in the data.

```python
# sns.boxplot(x='species', y='sepal_width', data=iris)
sns.boxplot(x='species', y='sepal_width', data=iris, hue='species')
```
