```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

**Line Graph**

```python
gas = pd.read_csv("gas_prices.csv")

plt.figure(figsize=(8,5))

plt.title("Gas Prices Over Time")
plt.xlabel("Year")
plt.ylabel("Dollars/Gallon")

plt.plot(gas.Year, gas.USA, 'r.-', label="USA")
plt.plot(gas.Year, gas['Germany'], 'b.-', label="Germany")
plt.plot(gas.Year, gas['South Korea'], 'g.-', label="South Korea")
plt.plot(gas.Year, gas.UK, 'y.-', label="UK")

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

```python
gas = pd.read_csv("gas_prices.csv")

plt.figure(figsize=(8,5))

plt.title("Gas Prices Over Time", fontdict={'fontweight':'bold', 'fontsize':'16'})
plt.xlabel("Year")
plt.ylabel("Dollars/Gallon")

for country in gas[1:]:
    if(country != 'Year'):
        plt.plot(gas.Year, gas[country], '.-', label=country)

plt.xticks(gas.Year[::3])

plt.legend()

plt.savefig("Gas_Price_Graph", dpi=300)

plt.show()
```

```python
gas = pd.read_csv("gas_prices.csv")

plt.figure(figsize=(8,5))

plt.title("Gas Prices Over Time")
plt.xlabel("Year")
plt.ylabel("Dollars/Gallon")

countries = ['Italy', 'Mexico', 'France', 'Canada']

for country in countries:
    if country in countries:
        plt.plot(gas.Year, gas[country], '.-', label=country)

plt.xticks(gas.Year[::3].tolist() + [2011])

plt.legend()

plt.show()
```
