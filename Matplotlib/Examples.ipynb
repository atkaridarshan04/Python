# Gas Price Example

**Line Graph**

```python
gas = pd.read_csv("gas_prices.csv")

plt.figure(figsize=(8,5))

plt.title("Gas Prices Over Time")
plt.xlabel("Year")
plt.ylabel("Dollars/Gallon")

plt.plot(gas.Year, gas.USA, 'r.-', label="USA")
plt.plot(gas.Year, gas['Germany'], 'b.-', label="Germany")
plt.plot(gas.Year, gas['South Korea'], 'g.-', label="South Korea")
plt.plot(gas.Year, gas.UK, 'y.-', label="UK")

plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```

```python
gas = pd.read_csv("gas_prices.csv")

plt.figure(figsize=(8,5))

plt.title("Gas Prices Over Time", fontdict={'fontweight':'bold', 'fontsize':'16'})
plt.xlabel("Year")
plt.ylabel("Dollars/Gallon")

for country in gas[1:]:
    if(country != 'Year'):
        plt.plot(gas.Year, gas[country], '.-', label=country)

plt.xticks(gas.Year[::3])

plt.legend()

plt.savefig("Gas_Price_Graph", dpi=300)

plt.show()
```

```python
gas = pd.read_csv("gas_prices.csv")

plt.figure(figsize=(8,5))

plt.title("Gas Prices Over Time")
plt.xlabel("Year")
plt.ylabel("Dollars/Gallon")

countries = ['Italy', 'Mexico', 'France', 'Canada']

for country in countries:
    if country in countries:
        plt.plot(gas.Year, gas[country], '.-', label=country)

plt.xticks(gas.Year[::3].tolist() + [2011])

plt.legend()

plt.show()
```

### Stack Plot

```python
yesr = gas['Year']

Australia   = gas['Australia'].tolist()
Mexico   = gas ['Mexico'].tolist()
USA = gas ['USA'].tolist()
UK   = gas ['UK'].tolist()
Canada   = gas['Canada'].tolist()
Japan = gas ['Japan'].tolist()

plt.plot([],[],color='m', label='Australia', linewidth=5)
plt.plot([],[],color='c', label='Mexico', linewidth=5)
plt.plot([],[],color='r', label='USA', linewidth=5)
plt.plot([],[],color='k', label='UK', linewidth=5)
plt.plot([],[],color='g', label='Canada', linewidth=5)
plt.plot([],[],color='y', label='Japan', linewidth=5)

plt.stackplot(year, Australia, Mexico, USA,UK, Canada,Japan, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Year ')
plt.ylabel('Dollars')
plt.title('Gas Price Track using Stack Plot')

plt.legend(loc='upper left')
plt.show()
```

# Fifa Data Example

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head()
```

### Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.title("Distribution of Player Skills in FIFA 2018")
plt.xlabel("Skill Level")
plt.ylabel("Number of Players")

plt.xticks(bins)

plt.show()
```

### Pie Chart

```python
fifa.head()
```

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count().iloc[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count().iloc[0]

plt.title("Foot Prefrence of FIFA Players")

labels = ["Left", "Right"]
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors = colors, autopct = '%.2f %%')

plt.show()
```

```python
plt.figure(figsize=(8, 5), dpi=100)
plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count().iloc[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count().iloc[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count().iloc[0]
heavy = fifa[fifa.Weight >= 200].count().iloc[0]

weights = [light, light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4, .2, 0, 0, .4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8, autopct='%.2f %%')

plt.show()
```

### Box and Whiskers Chart

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
