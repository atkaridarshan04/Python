```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Line Graph

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x,y)

# Add a title (specify parameters with fontdict)
plt.title("First Graph", fontdict={'fontname':'Comic Sans MS', 'fontsize' : 20})

# X and Y labels
plt.xlabel('X Axis', fontdict={'fontname':'Comic Sans MS'})
plt.ylabel('Y Axis')

# plt.show() # try running removing this
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Keyword Argument Notation
# plt.plot(x,y, label='2x', color='red', linestyle=':', linewidth='2', marker='.', markersize=10, markeredgecolor='darkblue')

# shorthand notation -> fmt = '[color][marker][line]'
plt.plot(x,y, 'kD--', label='2x', markerfacecolor='red')

plt.title("Second Graph")

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, 'bo--', label='2x')

plt.title("Third Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Second Line
# select interval we want to plot point at
x2 = np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2, x2**2, 'rH-', label='x^2', alpha=0.5)
# alpha sets transparency

# Add a legend
plt.legend()

# show plot
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resize Graph
plt.figure(figsize=(5,3), dpi=100)

plt.plot(x,y, 'bo--', label='2x')

plt.title("Third Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Second Line
x2 = np.arange(0,4.5,0.5)

# Plot part of graph as line
plt.plot(x2[:6], x2[:6]**2, 'rH-', label='x^2')
# Plot remainder of graph
plt.plot(x2[5:], x2[5:]**2, 'gh-', label='x^2' )

plt.legend(loc=6)

# Save figure (dpi = 300 is good when saving as a graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

plt.show()
```

### Other 2D Graphs

```python
n = np.array([1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize = (16,5))

axes[0].set_title("Scatter Graph")
axes[0].scatter(n, n**2, color='red')

axes[1].set_title("Step")
axes[1].step(n, n**2, 'y')

axes[2].set_title("Bar")
axes[2].bar(n, n ** 2, align='center', width=0.8)

axes[3].set_title("Bar")
axes[3].fill_between(n, n ** 2, n**3 ,color='g', alpha=0.5)

plt.show()
```

```python
# Radar Chart
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0, 0.0, 1, 0.6], polar=True)

t = np.linspace(0, 2*np.pi, 100)
ax.plot(t, t*10, color='blue', lw=2)

plt.show()
```

## Histograms

```python
n = np.random.randn(10000)
fig, axes = plt.subplots(1,2, figsize=(12,4))

axes[0].set_title("Default Histogram")
axes[0].hist(n, color='g', alpha=0.6)

axes[1].set_title("Cumulative Detailed Histogram")
axes[1].hist(n, cumulative=True, bins=50, color='blue', alpha=0.6)

plt.show()
```

# Pie Chart

```python
# Data to plot
languages = 'Python', 'Java', 'PHP', 'JavaScript', 'C#', 'C++'
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.style.use('ggplot')
plt.title("Popularity of Programming Languages ")

label = ['Python', 'Java', 'PHP', 'JavaScript', 'C#', 'C++']
explode =[0.05, 0., 0, 0, 0, 0]

plt.pie(popularity, labels = label, explode = explode, autopct = '%1.1f%%', startangle=90)

plt.show()
```

# Contour Image

```python
import matplotlib.cm as cm # cm = color map

delta = 0.25
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)

X, Y = np.meshgrid(x, y)
z1 = np.exp(-X**2 - Y**2) # exp = exponential
z2 = np.exp( - (-X-1)**2 - (Y-1)**2)
Z = (z1 - z2) * 2

# print(X)
# print(Y)
# print(Z)
```

```python
fig, axes = plt.subplots()
CS = axes.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)

plt.show()
```
